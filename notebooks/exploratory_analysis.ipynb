# ML Mirchi — Exploratory Data Analysis

Use this notebook to explore your dataset before training.

```python
import os
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

DATA_DIR = Path("../data/processed")
CLASSES = ["Grade_A", "Grade_B", "Grade_C", "Reject"]
```

```python
# Count images per class per split
for split in ["train", "val", "test"]:
    print(f"\n{split.upper()}:")
    for cls in CLASSES:
        cls_dir = DATA_DIR / split / cls
        count = len(list(cls_dir.glob("*.jpg"))) + len(list(cls_dir.glob("*.png")))
        print(f"  {cls}: {count}")
```

```python
# Sample images from each class
fig, axes = plt.subplots(4, 5, figsize=(15, 12))
for row, cls in enumerate(CLASSES):
    cls_dir = DATA_DIR / "train" / cls
    images = list(cls_dir.glob("*.jpg"))[:5]
    for col, img_path in enumerate(images):
        img = Image.open(img_path)
        axes[row, col].imshow(img)
        axes[row, col].axis("off")
        if col == 0:
            axes[row, col].set_ylabel(cls, fontsize=12, fontweight="bold")
plt.suptitle("Sample Images from Each Grade", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# Color distribution analysis
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i, cls in enumerate(CLASSES):
    cls_dir = DATA_DIR / "train" / cls
    reds, greens, blues = [], [], []
    for img_path in list(cls_dir.glob("*.jpg"))[:50]:
        img = np.array(Image.open(img_path))
        reds.append(img[:,:,0].mean())
        greens.append(img[:,:,1].mean())
        blues.append(img[:,:,2].mean())
    axes[i].hist(reds, bins=20, alpha=0.7, color='red', label='R')
    axes[i].hist(greens, bins=20, alpha=0.7, color='green', label='G')
    axes[i].hist(blues, bins=20, alpha=0.7, color='blue', label='B')
    axes[i].set_title(cls)
    axes[i].legend(fontsize=8)
plt.suptitle("Color Channel Distribution by Grade", fontsize=14)
plt.tight_layout()
plt.show()
```
